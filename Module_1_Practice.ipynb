```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("companies.csv")
```

```python
df.head()
```

```python
# question 1
plt.figure (figsize = (8,5))
sns.countplot(x= "state",data = df)
plt.show()
```

```python
plt.figure(figsize = (8,5))
sns.barplot(y = 'Industry',x = 'job_openings',data = df)
plt.show()
```

```python
plt.figure(figsize = (8,5))
sns.lineplot(x = 'founded',y = 'growth_percentage',data = df)
plt.show()
```

```python
agg_df = df [['founded','growth_percentage']].groupby('founded').mean().reset_index()
agg_df.head()
```

```python
plt.figure(figsize = (8,5))
sns.scatterplot(x = 'employees',y = 'estimated_revenues',data = df)
plt.show()
```

```python
sns.histplot(x = 'founded',data = df)
plt.show()
```
